```python
import pandas as pd
import matplotlib.pyplot as plt

FILEPATH = "data/unknown_data.csv"

df = pd.read_csv(FILEPATH, header=None)
df.describe()
```

```python
fig, ax = plt.subplots()
x = range(df.shape[1])
ax.stackplot(x, df)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.imshow(df)
plt.axis("off")
plt.colorbar(label="Intensity")

plt.show()
```
