```python
import pandas as pd

# Define the data
data = {
    "Year": [2019, 2020, 2021, 2022, 2023],
    "Company A": [500, 520, 540, 580, 620],
    "Company B": [550, 580, 610, 640, 690],
    "Company C": [530, 540, 550, 570, 600],
    "Company D": [580, 600, 630, 670, 710],
    "Company E": [620, 650, 680, 720, 760],
}
```

```python
# Create the DataFrame and format
df = pd.DataFrame(data)
df["Year"] = pd.to_datetime(df["Year"], format="%Y")
df.set_index("Year", inplace=True)
```

```python
import matplotlib.pyplot as plt

plt.plot(df.iloc[:, :-1])
plt.show()
```

```python
plt.plot(df.iloc[:, :-1])

plt.title("Company Revenues")

plt.legend(df.iloc[:, :-1])

plt.xlabel("Year")
years = df.index.unique()
plt.xticks(years, [x.year for x in years], rotation=60)

plt.ylabel("Revenue")

plt.grid()

plt.show()
```
