```python
import pandas as pd

FILEPATH = "data/marketing_vs_revenue.csv"

df = pd.read_csv(FILEPATH)
df.set_index("Marketing Spend", inplace=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math

plt.scatter(df.index, df.iloc[:, :])

plt.title("Marketing Spend vs Revenue")

plt.xlabel("Marketing Spend")
min_x = math.floor(df.index.min())
max_x = math.ceil(df.index.max())

step_val_x = math.ceil((max_x - min_x) / 10)
plt.xticks(np.arange(min_x - step_val_x, max_x + step_val_x, step_val_x), rotation=60)

plt.ylabel("Revenue")

m, b = np.polyfit(df.index.dropna(), df.iloc[:, :].dropna(), 1, cov=True)[0]
plt.plot(df.index, m * df.index + b, color="red")

plt.legend(["Data Points", f"Trendline y = {m[0]:.2f}x + {b[0]:.2f}"])

plt.show()
```
